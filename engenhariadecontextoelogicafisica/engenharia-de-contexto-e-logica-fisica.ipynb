```python
### PARTE 1: O Algoritmo do Microclima Local

def microclima_local():

  print("=== Análise de Microclima Local ===")



  # Lista de listas (dados coletados)

  dados_locais = [

    ["Itaquaquecetuba - manha", 25, 41, 57],

    ["Itaquaquecetuba - tarde", 23, 41, 57],

    ["Guarulhos - manha", 21, 68, 73],

    ["Guarulhos - tarde", 24, 56, 73],

    ["Carrão - manha", 25, 52, 77],

    ["Carrão - tarde", 27, 48, 77]

  ]



  # FOR principal

  for local in dados_locais:

    nome, temperatura, umidade, iaq = local



    print(f"\nLocal: {nome}")

    print(f"Temperatura: {temperatura}°C")

    print(f"Umidade: {umidade}%")

    print(f"IAQ: {iaq}")



    # Chamando funções

    classificacao = classificar_iaq(iaq)

    nota = calcular_conforto(temperatura, umidade, iaq)



    # Exibição dos resultados

    print(f"Classificação IAQ: {classificacao}")

    print(f"Nota de conforto: {nota}")





# Função para classificar o IAQ (match-case)

def classificar_iaq(iaq):

  match iaq:

    case i if i <= 50:

      return "Boa - qualidade satisfatória, risco baixo ou nulo"

    case i if i <= 100:

      return "Moderada - qualidade aceitável, risco para pessoas sensíveis"

    case i if i <= 150:

      return "Insalubre para grupos de risco"

    case i if i <= 200:

      return "Insalubre"

    case i if i <= 300:

      return "Muito insalubre"

    case _:

      return "Extremamente insalubre"





# Função para calcular a nota de conforto

def calcular_conforto(temperatura, umidade, iaq):

  penal_temperatura = abs(temperatura - 20) / 2

  penal_umidade = abs(umidade - 40) / 4

  penal_iaq = iaq / 50



  nota = 10 - (penal_temperatura + penal_umidade + penal_iaq)



  # Garantir que a nota fique entre 0 e 10

  if nota < 0:

    nota = 0

  elif nota > 10:

    nota = 10



  return round(nota, 2)





# Execução do programa

microclima_local()





### PARTE 2: Navegação e Evacuação Espacial

# 1. Definição das Listas (Matriz de obstáculos)

locais = ["SALA", "CORREDOR", "ESCADA", "SAIDA"]



# Cada sub-lista [] dentro da lista principal representa os nós de cada local

mapa_obstaculos = [

  ["MESA", "CADEIRAS", "PORTA"],     # Nós da Sala

  ["EXTINTOR", "TAPETES", "PORTA"],    # Nós do Corredor

  ["LANCE DE ESCADA", "EXTINTOR", "PORTA"],# Nós da Escada

  ["LIVRE"]                # Saída

]



# Estado do Agente

posicao_atual = 0 # Índice da lista 'locais'

inventario = []  # Onde guardaremos itens encontrados

energia = 15    # Limite de ciclos do loop



print("--- INICIANDO EVACUAÇÃO ---")



while posicao_atual < len(locais) and energia > 0:

  local = locais[posicao_atual]

  nos_atuais = mapa_obstaculos[posicao_atual]



  print(f"\nLocal atual: {local}")



  # Lógica com Ifs Aninhados para avaliar os nós (obstáculos)

  if local == "SALA":

    print(f"Observando: {nos_atuais}")

    # Se encontrar a porta, tenta avançar

    if "PORTA" in nos_atuais:

      print("Porta da sala aberta. Avançando para o corredor.")

      inventario.append("CHAVE_MESTRA") # Agente pegou algo útil

      posicao_atual = posicao_atual + 1



  elif local == "CORREDOR":

    if "TAPETES" in nos_atuais:

      print("Cuidado com os tapetes escorregadios!")

      if "CHAVE_MESTRA" in inventario:

        print("Usando chave encontrada na sala para abrir a porta do corredor.")

        posicao_atual = posicao_atual + 1

      else:

        print("Porta trancada e sem chave! Voltando para procurar.")

        posicao_atual = posicao_atual - 1



  elif local == "ESCADA":

    if "LANCE DE ESCADA" in nos_atuais:

      print("Descendo o lance de escada...")

      if "EXTINTOR" in nos_atuais:

        print("Vi um extintor, mas o caminho está livre. Indo para a saída.")

        posicao_atual = posicao_atual + 1



  elif local == "SAIDA":

    break # Sai do loop imediatamente



  energia = energia - 1



# Resultado Final

if posicao_atual >= 3:

  print("\n[SUCESSO] O agente alcançou a RUA!")

else:

  print("\n[ERRO] O agente falhou na evacuação.")
```
