```
# Sistema Inteligente de Controle de Semáforos
# Tema: Cidades Inteligentes

while True:

    print("\n=== CONTROLE INTELIGENTE DE SEMÁFOROS ===")

    # Entrada de dados
    carrosA = int(input("Quantidade de carros na Rua A: "))
    carrosB = int(input("Quantidade de carros na Rua B: "))

    # Verificação de erro
    if carrosA < 0 or carrosB < 0:

        print("ERRO: Quantidade de veículos inválida.")

    else:

        # Decisão 1
        if carrosA > carrosB:

            tempoA = 60
            tempoB = 30

            print("\nRua A possui maior fluxo.")
            print(f"Sinal verde Rua A: {tempoA} segundos")
            print(f"Sinal verde Rua B: {tempoB} segundos")

        # Decisão 2
        elif carrosB > carrosA:

            tempoA = 30
            tempoB = 60

            print("\nRua B possui maior fluxo.")
            print(f"Sinal verde Rua A: {tempoA} segundos")
            print(f"Sinal verde Rua B: {tempoB} segundos")

        # Decisão 3
        else:

            tempoA = 45
            tempoB = 45

            print("\nFluxo equilibrado.")
            print(f"Sinal verde Rua A: {tempoA} segundos")
            print(f"Sinal verde Rua B: {tempoB} segundos")

    # Pergunta para continuar
    continuar = input("\nDeseja realizar nova análise? (s/n): ")

    if continuar.lower() != "s":
        print("\nSistema encerrado.")
        break
```
