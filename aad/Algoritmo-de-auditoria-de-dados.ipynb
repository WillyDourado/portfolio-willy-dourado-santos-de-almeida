O Algoritmo de Auditoria de Dados

```
# Auditoria de Vendas Semanais

# Configuração Global
LIMITE_SEGURANCA = 10000.0  # Valor máximo sem ser suspeito

def analisar_vendas(v1, v2, v3):
    global LIMITE_SEGURANCA  # Permite alterar o limite se necessário

    vendas = [v1, v2, v3]

    # Cálculo da média simples
    media = sum(vendas) / len(vendas)

    print(f"\nMédia das vendas: {media} ({type(media)})")

    # Lógica de investigação
    if media > LIMITE_SEGURANCA:
        print("ALERTA: SISTEMA EM QUARENTENA!")

    for i, venda in enumerate(vendas, start=1):
        if venda > 5 * media:
            print(f"Venda {i} = {venda} é 5x maior que a média! REVISÃO MANUAL SUGERIDA.")

    # Exibição dos tipos de dados das vendas
    for i, venda in enumerate(vendas, start=1):
        print(f"Venda {i}: {venda} ({type(venda)})")

    # Permite ao gerente ajustar o limite
    resposta = input("\nDeseja ajustar o LIMITE_SEGURANCA? (s/n): ")
    if resposta.lower() == 's':
        novo_limite = float(input("Digite o novo LIMITE_SEGURANCA: "))
        LIMITE_SEGURANCA = novo_limite
        print(f"Novo LIMITE_SEGURANCA definido para {LIMITE_SEGURANCA}")

# Entrada de dados do usuário
venda1 = float(input("Digite o valor da venda 1: "))
venda2 = float(input("Digite o valor da venda 2: "))
venda3 = float(input("Digite o valor da venda 3: "))

# Chamando a função
analisar_vendas(venda1, venda2, venda3)

# Exibindo valor global atualizado
print(f"\nLIMITE_SEGURANCA atual: {LIMITE_SEGURANCA} ({type(LIMITE_SEGURANCA)})")
```
